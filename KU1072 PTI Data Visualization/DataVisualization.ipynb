# Import Library & Configure Pandas Settings

```python
import pandas as pd
import matplotlib.pyplot as plt

desired_width = 128
pd.set_option('display.width', desired_width)
pd.set_option('display.max_columns',16)
pd.set_option('display.max_rows',8)
```

# Convert Raw Data (CSV) to Pandas Dataframe, Data Cleansing

```python
#Meload data file menjadi dataframe dengan variabel df
dfkotor = pd.read_csv("processed-penduduk-miskin-indeks-kemiskinan-2007-2010.csv")
print("Data Kotor Memiliki nilai Null Sebanyak: ")
print(dfkotor.isnull().sum())

# membersihkan data kosong/null
df = dfkotor.dropna() 
print("\nNilai Null Setiap Kolom Data Setelah Dibersihkan: ")
print(df.isnull().sum())
```

# Data Visualization

```python
# Memisahkan Dataframe untuk masing masing tahun
df2007 = df.loc[df["tahun"] == 2007]
df2008 = df.loc[df["tahun"] == 2008]
df2009 = df.loc[df["tahun"] == 2009]
df2010 = df.loc[df["tahun"] == 2010]
```

## Showing Over Time

```python
# jumlah penduduk miskin di indonesia tahun 2007-2010 (setiap tahun)
jpm_2007 = df2007["jumlah_penduduk_miskin"].sum() #37062700
jpm_2008 = df2008["jumlah_penduduk_miskin"].sum() #34543900
jpm_2009 = df2009["jumlah_penduduk_miskin"].sum() #31772500
jpm_2010 = df2010["jumlah_penduduk_miskin"].sum() #31065300

# membuat dataframe baru rata-rata jpm (jumlah penduduk miskin) untuk dibuat grafiknya
dictjpm = {"jumlah_penduduk_miskin" : pd.Series([jpm_2007,jpm_2008,jpm_2009,jpm_2010]),
           "tahun" : pd.Series([2007,2008,2009,2010])}
dfjpm = pd.DataFrame(dictjpm)
dfjpm.plot(kind = 'line',
           x = "tahun",
           y = "jumlah_penduduk_miskin", 
           title = "Jumlah Penduduk Miskin di Indonesia Thn 2007-2010",
           xticks=[2007,2008,2009,2010],
           figsize=(8,6))
plt.show()

```

```python
#rata-rata garis kemiskinan di Indonesia tahun 2007-2010(setiap tahun)
gk_2007 = (df2007["garis_kemiskinan"].mean()) #175089
gk_2008 = (df2008["garis_kemiskinan"].mean()) #198447
gk_2009 = (df2009["garis_kemiskinan"].mean()) #223211
gk_2010 = (df2010["garis_kemiskinan"].mean()) #245546

#membuat dataframe baru rata-rata gk (garis kemiskinan) untuk dibuat grafiknya
dictgk = {"garis_kemiskinan" : pd.Series([gk_2007,gk_2008,gk_2009,gk_2010]),
           "tahun" : pd.Series([2007,2008,2009,2010])}
dfgk = pd.DataFrame(dictgk)
dfgk.plot(kind = 'line',
           x = "tahun",
           y = "garis_kemiskinan", 
           title = "Rata-rata Garis Kemiskinan di Indonesia Thn 2007-2010",
           xticks=[2007,2008,2009,2010],
           figsize=(8,6))
plt.show()
""""""
```

```python
#rata-rata Poverty Gap Index (indeks_p1)di indonesia tahun 2007-2010 (setiap tahun)
p1_2007 = (df2007["indeks_p1"].mean()) #3.228
p1_2008 = (df2008["indeks_p1"].mean()) #4.003
p1_2009 = (df2009["indeks_p1"].mean()) #2.853
p1_2010 = (df2010["indeks_p1"].mean()) #2.813

#membuat dataframe baru rata-rata indeks_p1(poverty gap index) untuk dibuat grafiknya
dictp1 = {"indeks_p1" : pd.Series([p1_2007,p1_2008,p1_2009,p1_2010]),
           "tahun" : pd.Series([2007,2008,2009,2010])}
dfp1 = pd.DataFrame(dictp1)
dfp1.plot(kind = 'line',
           x = "tahun",
           y = "indeks_p1", 
           title = "Rata-rata Poverty Gap Index di Indonesia Thn 2007-2010",
           xticks=[2007,2008,2009,2010],
           figsize=(8,6))
plt.show()
```

## Whole-Part Relationship

### Dataframe Splitting Into 3 Parts of Indonesia (West, Middle, East)

```python
dfbarat = df.loc[((df["kode_provinsi"] < 51) | (df["kode_provinsi"] == 62) | (df["kode_provinsi"] == 61))]
dftengah = df.loc[((df["kode_provinsi"] >= 51) & (df["kode_provinsi"] <= 76)) & (df["kode_provinsi"] != 62) & (df["kode_provinsi"] != 61)]
dftimur = df.loc[df["kode_provinsi"] > 76]
```

### Year 2007

```python
# 2007
dfbarat2007 = dfbarat.loc[df["tahun"] == 2007]
dftengah2007 = dftengah.loc[df["tahun"] == 2007]
dftimur2007 = dftimur.loc[df["tahun"] == 2007]
a = dfbarat2007["persen_penduduk_miskin"].mean()
b = dftengah2007["persen_penduduk_miskin"].mean()
c = dftimur2007["persen_penduduk_miskin"].mean()
#persen_penduduk_miskin
#barat                16.537626
#tengah               17.824394
#timur                33.662558
#avg                  68.024578

dictppm = {"persen_penduduk_miskin" : pd.Series([a,b,c], index = ["barat","tengah","timur"])}
dfppm = pd.DataFrame(dictppm)
print(dfppm)
dfppm.plot(kind = "pie",subplots = True,title = "Persen penduduk miskin pada 3 wilayah Indonesia (2007, total 68.02%)")

labels = [r'barat (16.53%)', r'tengah (17.82%)', 
r'timur (33.66%)']
sizes = [24.32, 26.2, 49.48]
colors = ['yellowgreen', 'lightskyblue', 'lightcoral']
patches, texts = plt.pie(sizes, colors=colors, startangle=0)
plt.legend(patches,labels, bbox_to_anchor=(0.85,0.05), loc="lower left")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
